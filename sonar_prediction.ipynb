```python
#diving our data
x=db.drop(columns=[60],axis=1)
y=db[60]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.1,random_state=1,stratify=y)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model_predicition=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
test_accuracy=accuracy_score(model_predicition,y_test)
print('accuracy of our model is = ',test_accuracy)
```
